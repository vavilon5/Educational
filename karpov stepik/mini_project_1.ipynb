```python
import pandas as pd
```

```python
df = pd.read_csv('bookings.csv', sep=';')
df.head(7)
```

```python
df.shape
```

```python
df.dtypes
```

```python
renamed = dict()

for col in df.columns:
    new_col = '_'.join(col.split()).lower()
    renamed[col] =  new_col
    
df = df.rename(columns = renamed)
df.head()
```

### Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
df.query('is_canceled == 0') \
  .groupby('country') \
  .agg({'is_canceled': 'count'})\
  .sort_values(by='is_canceled', ascending=False)\
  .head()
```

### На сколько ночей в среднем бронируют отели разных типов?

```python
df.groupby('hotel').agg({'stays_total_nights' : 'mean'})
```

```python
df['hotel'].unique()
```

### Сколько забранированных номеров отличаются от полученных?

```python
df.query('assigned_room_type != reserved_room_type').agg({'assigned_room_type' : 'count'})
```

### На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?

```python
df.query('(arrival_date_year == 2016) and (is_canceled == 0)')\
  .groupby('arrival_date_month', as_index=False)\
  .agg({'lead_time': 'count'})\
  .sort_values(by='lead_time', ascending=False).head(1)

# October
```

```python
df.query('(arrival_date_year == 2017) and (is_canceled == 0)')\
  .groupby('arrival_date_month', as_index=False)\
  .agg({'lead_time': 'count'})\
  .sort_values(by='lead_time', ascending=False).head(1)
# May
```

### Посмотрите на числовые характеристики трёх переменных: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
print(df['adults'].mean(),
df['children'].mean(),
df['babies'].mean())
# adults
```

### Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

```python
df['total_kids'] = df['children'] + df['babies']
```

```python
df.groupby('hotel').agg({'total_kids': 'mean'})
```

###  Среди какой группы пользователей (с / без детей) показатель оттока выше

```python
# df = df.drop(columns='has_kids')
df['has_kids'] = False
for i in df.index:
    df['has_kids'][i] = bool(df['total_kids'][i])
        
df
```

```python
df.groupby('has_kids', as_index=False).agg({'is_canceled': ['count', 'sum']})
```
