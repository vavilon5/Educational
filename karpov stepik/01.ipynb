```python
import pandas as pd
```

```python
df = pd.read_csv('lesson_1_data.csv', encoding='windows-1251', sep=';')
```

```python
df
```

```python
#df.columns

renamed = {'Номер': 'number', 
           'Дата создания': 'create_date', 
           'Дата оплаты': 'payment_date', 
           'Title': 'title', 
           'Статус': 'status',
           'Заработано': 'money', 
           'Город' : 'city', 
           'Платежная система': 'payment_system'}
df = df.rename(columns=renamed)
```

```python
df
```

```python
df.title.unique()
```

```python
df['title']
```

```python
all_money = round(df.money.sum(), 2)

all_money
```

```python
 money_by_city = df \
    .groupby(['title', 'city'], as_index=False) \
    .aggregate({'money': 'sum'}) \
    .sort_values('money', ascending=False)
```

```python
money_by_city.to_csv('money_by_city.csv', index=False)
```

```python
money_title = df \
    .query('status == "Завершен"') \
    .groupby(['title'], as_index=False) \
    .aggregate({'money': 'sum', 'number': 'count'}) \
    .sort_values('money', ascending=False) \
    .rename(columns={'number': 'success_orders'})

money_title
```

```python
if int(money_title.money.sum) == int(all_money):
    print('good')
else:
    print('NOOOT GOOOOOD')
```

```python
taxi = pd.read_csv('2_taxi_nyc.csv')

taxi
```

```python
renamed = dict()
for item in taxi.columns:
    if ' ' in item:
        cur = item.replace(' ', '_')
        renamed[item] = cur

renamed
```

```python
taxi = taxi.rename(columns=renamed)
taxi.head()
```

```python
taxi['borough'].value_counts()
```

```python
drives = taxi.groupby('borough', as_index=False).aggregate({'pickups': 'sum'}).sort_values('pickups', ascending=False)

#taxi.pickups.sum()
drives
```

```python
holidays = taxi.query('hday == "Y"').groupby('borough', as_index=False).aggregate({'pickups': 'mean'}).sort_values('pickups', ascending=False)

holidays
#meany
```

```python
workdays = taxi.query('hday == "N"').groupby('borough', as_index=False).aggregate({'pickups': 'mean'}).sort_values('pickups', ascending=False)
workdays
```
