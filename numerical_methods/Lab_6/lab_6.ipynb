$$f(x) = \, –cos(2x + 1)\, ln(2 / x) + 3$$

##Метод Золотого Сечения
$$x_1 = b - \frac{b-a}{𝚽}$$
$$x_2 = a + \frac{b-a}{𝚽}$$

$$𝚽 = \frac{1+\sqrt{5}}{2}$$

```python
n = 20
a = 1.5
b = 2.3
phi = (1 + np.sqrt(5)) / 2

df_gs = pd.DataFrame(columns=['a', 'b', 'x1', 'x2', 'y1', 'y2'])
delta = []

x1, x2 = b - (b - a) / phi, a + (b - a) / phi
y1, y2 = f(x1), f(x2)

for i in range(n):

  df_gs.loc[len(df_gs.index)] = [a, b, x1, x2, y1, y2]

  # Фишка ЗС, в том, что функцию можно вызывать один раз
  if y1 >= y2:
    a = x1
    x1 = x2
    x2 = a + (b - a) / phi
    y1 = y2
    y2 = f(x2)
  else:
    b = x2
    x2 = x1
    x1 = b - (b - a) / phi
    y2 = y1
    y1 = f(x1)

  delta.append(abs(a-b))

df_gs['delta'] = delta

print((a+b)/2)
df_gs
```

##Метод дихотомии

$$x_1 = \frac{b+a}{2} - δ$$
$$x_2 = \frac{b+a}{2} + δ$$

$$δ∈(0;\, \frac{b-a}{2}), \, δ < \frac{ε}{2} $$

```python
d = 0.00002
n = 15
a = 1.5
b = 2.3
i = 0
df_d = pd.DataFrame(columns=['a', 'b', 'x1', 'x2', 'y1', 'y2'])
delta = []


for _ in range(n):
  x1, x2 = (a + b) / 2 - d, (a + b) / 2 + d
  y1, y2 = f(x1), f(x2)

  df_d.loc[len(df_d.index)] = [a, b, x1, x2, y1, y2]

  if y1 > y2:
    a = x1
  else:
    b = x2

  delta.append(b - a)

print((x1 + x2) / 2)
```
