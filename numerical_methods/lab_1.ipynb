##Этап отделения корней

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

from sympy import *
```

```python
def f(x):
    return np.cos(x) - np.sqrt(x+2) + 1
```

```python
begin = -2
end = 7
step = 0.2
size = 1000
x = np.linspace(begin, end, size)
y_f = f(x)
```

```python
fig = go.Figure() # layout_xaxis_range = [-2,6]
fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_f,
                         name="График функции"))

fig.update_layout(title="Построение графика функции",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции",
                  margin=dict(l=10, r=70, t=50, b=0))

fig.show()
```

Из построенного графика функции $f(x)$ видно, что на отрезке $(0, 1.5)$ есть один корень.

```python
# Вычислим производную функции f(z)
z = Symbol('z')

f_ = cos(z) - sqrt(z+2) + 1
f_1derivative = f_.diff(z)

f1 = lambdify(z, f_1derivative)

f2 = lambdify(z, f_1derivative.diff(z))
```

```python
f_1derivative
```

```python
a = 0
b = 1.5
begin = a - 1
end = b + 1
size = 1000
x = np.linspace(begin, end, size)
```

```python
x_temp = np.arange(a, b+0.1, 0.1)
df = pd.DataFrame(data={'x': x_temp,
                        'f(x)': f(x_temp),
                        'f1(x)': f1(x_temp),
                        'f2(x)': f2(x_temp)})
df
```

##Этап уточнения корня

###Метод итераций

Приведем уравнение $f(x)=0$ к виду $x=𝞅(x)$. Тогда рекуррентная формула $x_{n+1}=𝞅(x_{n})$. Для сходимости процесса простых итерации необходимо, чтобы $|𝞅'(x)|<1$ при $x\in [a;b]$. Если $|𝞅'(x)|>1$ то сходимость не обеспечена.
Приведем уравнение $f(x)=cos(x)–\sqrt{x + 2} + 1$ к виду $x=(cos(x)+1)^2 - 2$ и проведем исследование.


```python
# Определим функцию ф
def phi(x):
    return ((np.cos(x)+1)**2 - 2)
```

```python
# Вычислим первую производную функции ф
z = Symbol('z')

phi_ = ((cos(z)+1)**2 - 2)

phi_derivative = phi_.diff(z)
phi_derivative
```

```python
phi1 = lambdify(z, phi_derivative)
```

```python
y_phi = phi(x)
y_phi1 = phi1(x)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_phi,
                         name="График функции фи"))

fig.add_trace(go.Scatter(x=[a, a],
                         y=[-2, 2],
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b],
                         y=[-2, 2],
                         name="Ось B"))

fig.add_trace(go.Scatter(x=x,
                         y=x,
                         name="Ограничение сверху"))

fig.update_layout(title="Построение функции фи",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

```python
# Изобразим график первой производной ф-ии ф
fig = go.Figure()
fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_phi1,
                         name="График производной функции фи"))

fig.add_trace(go.Scatter(x=[a, a],
                         y=[-2, 1],
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b],
                         y=[-2, 1],
                         name="Ось B"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[1, 1],
                         name="Ограничение сверху"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[-1, -1],
                         name="Ограничение снизу"))


fig.update_layout(title="Построение производной функции фи",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

Как видно на графике выше $|𝞅'(x)|>1$ на отрезке $(a,b)$, поэтому построим функцию $𝞅(x) = x + \lambda f(x)$, где параметр $\lambda$ может быть определен по правилу: $\lambda = -\frac{1}{f'(\hat{x})}$, а в знаменатель требуется подставить $f'(\hat{x})$ = $\max\limits_{a \leq x \leq b}|f'(x)|$. Найдем параметр $\lambda$.

```python
# phi1(x[(x > a) & (x < b)])
```

```python
# Вычислим производную функции f(z)
z = Symbol('z')

f_ = cos(z) - sqrt(z+2) + 1
f_1derivative = f_.diff(z)

f1 = lambdify(z, f_1derivative)
y_f1 = f1(x)

f_1derivative
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=x,
                         y=y_f1,
                         name="Первая производная функции"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))


fig.add_trace(go.Scatter(x=[a, a],
                         y=[-3, 2],
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b],
                         y=[-3, 2],
                         name="Ось B"))

fig.update_layout(title="Построение производной функции f",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

```python
# print(max(max(abs(y_f1)), min(y_f1)))
maxi = min(y_f1)
lamda = -1/maxi
lamda, maxi
```

Тем самым $\lambda=\frac{1}{1.265}$. Тогда рекурентная формула: $$𝞅(x)=x+\frac{cos(x)–\sqrt{x + 2} + 1}{1.265}$$

```python
def phi(x):
    return x + lamda * f(x)
```

```python
z = Symbol('z')

phi_ = z + lamda * f_

phi_derivative = phi_.diff(z)
phi_derivative
```

```python
phi1 = lambdify(z, phi_derivative)
```

```python
y_phi = phi(x)
y_phi1 = phi1(x)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_phi1,
                         name="График производной функции фи"))

fig.add_trace(go.Scatter(x=[a, a],
                         y=[-2, 1],
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b],
                         y=[-2, 1],
                         name="Ось B"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[1, 1],
                         name="Ограничение сверху"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[-1, -1],
                         name="Ограничение снизу"))


fig.update_layout(title="Построение производной функции фи",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

```python
x0 = a
sp_x = [x0]
sp_f = [f(x0)]
for i in range(3):
  x_n = phi(x0)
  sp_x.append(x_n)
  sp_f.append(f(x_n))
  x0 = x_n
```

```python
df = pd.DataFrame(data={'n':range(i+2), 'x': sp_x, 'f(x)': sp_f})
df
```

```python
q = max(phi1(x[(x>a)&(x<b)]))
R = q/(1-q) * abs(sp_x[-1] - sp_x[-2])
R
```

## Метод Ньютона

```python
# Вычислим производные функции f(z)
z = Symbol('z')

f_ = cos(z) - sqrt(z+2) + 1
f_1derivative = f_.diff(z)

f1 = lambdify(z, f_1derivative)
y_f1 = f1(x)

f_1derivative
# dx = x[1]-x[0]
# dydx = np.gradient(y, dx)
```

```python
f2 = lambdify(z, f_1derivative.diff(z))
y_f2 = f2(x)
f_1derivative.diff(z)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=x,
                         y=y_f1,
                         name="Первая производная функции"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))


fig.add_trace(go.Scatter(x=[a, a],
                         y=[-3, 2],
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b],
                         y=[-3, 2],
                         name="Ось B"))

fig.update_layout(title="Построение производной функции f",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

```python
fig = go.Figure(layout_yaxis_range=[-4,4])

fig.add_trace(go.Scatter(x=x,
                         y=y_f2,
                         name="Вторая производная функции"))

fig.add_trace(go.Scatter(x=[begin, end],
                         y=[0, 0],
                         name="Ось Х"))


fig.add_trace(go.Scatter(x=[a, a],
                         y=[-10, 2],
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b],
                         y=[-10, 2],
                         name="Ось B"))

fig.update_layout(title="Построение второй производной функции f",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")


fig.show()
```

```python
E = [i for i in range(-2, -5, -1)]

sp_i = []
for eps in E:
  x0 = a
  i = 0
  sp_x = [x0]
  sp_f = [f(x0)]
  x_n = x0 - f(x0)/f1(x0)
  while abs(x_n - x0) > 10**eps:
    if eps == -4:
      sp_x.append(x_n)
      sp_f.append(f(x_n))
    x0 = x_n
    x_n = x0 - f(x0)/f1(x0)
    i += 1
  sp_i.append(i)
```

```python
df = pd.DataFrame(data={'n':range(sp_i[-1]+1), 'x': sp_x, 'f(x)': sp_f})
df
```

```python
fig = go.Figure(layout_yaxis_range = [2,5])

fig.add_trace(go.Scatter(x=E,
                         y=sp_i,
                         name="Количество итераций для погрешности"))

fig.update_layout(title="Построение зависимости итерации от погрешности",
                  xaxis_title="Значение погрешности под логарифмом",
                  yaxis_title="Значение итерации")

fig.show()
```
