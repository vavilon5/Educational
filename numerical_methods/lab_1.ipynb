```python
import pandas as pd
import numpy as np
from sympy import *

import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
def func(x):
    return np.cos(x) - np.sqrt(x+2) + 1
```

```python
begin = -3
end = 10
step = 0.2
size = 1000
x = np.linspace(begin, end, size)
y_f = func(x)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[begin, end], 
                         y=[0, 0], 
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_f,
                         name="График функции"))

fig.update_layout(title="Построение графика функции",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

## Отрезок [0, 2]

```python
a = 0
b = 2
```

```python
z = Symbol('z')

f = cos(z) - sqrt(z+2) + 1
f_1derivative = f.diff(z)
f_1derivative
```

```python
f1 = lambdify(z, f_1derivative)
y_f1 = f1(x)

# dx = x[1]-x[0]
# dydx = np.gradient(y, dx)
```

```python
f2 = lambdify(z, f_1derivative.diff(z))
y_f2 = f2(x)
```

```python
f_1derivative.diff(z)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[begin, end], 
                         y=[0, 0], 
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_f1,
                         name="График производной функции"))

fig.add_trace(go.Scatter(x=[a, a], 
                         y=[-10, 1], 
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b], 
                         y=[-10, 1], 
                         name="Ось B"))


fig.update_layout(title="Построение производной функции",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```

## Метод простых итераций

```python
def phi(x):
    return (np.cos(x)+1)**2 - 2
```

```python
z = Symbol('z')

f = (cos(z)+1)**2 - 2
f_derivative = f.diff(z)
f_derivative
```

```python
phi1 = lambdify(z, f_derivative)
```

```python
y_phi = phi(x)
y_phi1 = phi1(x)
```

```python
x
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[begin, end], 
                         y=[0, 0], 
                         name="Ось Х"))

fig.add_trace(go.Scatter(x=x,
                         y=y_phi1,
                         name="График производной функции фи"))

fig.add_trace(go.Scatter(x=[a, a], 
                         y=[-10, 1], 
                         name="Ось A"))

fig.add_trace(go.Scatter(x=[b, b], 
                         y=[-10, 1], 
                         name="Ось B"))

fig.add_trace(go.Scatter(x=[begin, end], 
                         y=[1, 1], 
                         name="Ограничение сверху"))

fig.add_trace(go.Scatter(x=[begin, end], 
                         y=[-1, -1], 
                         name="Ограничение снизу"))


fig.update_layout(title="Построение производной функции фи",
                  xaxis_title="Значение аргумента",
                  yaxis_title="Значение функции")

fig.show()
```
