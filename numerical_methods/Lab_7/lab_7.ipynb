```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

from sympy import *
```

```python
# Определяем многомерную функцию
def f(x, y):
  return 5 * x**2 + 6 * y**2 + 3*x - 2*y + 3
```

```python
# Точное значение функции
f(-0.3, 1/6)
```

```python
# Считаем производную по l(lambda)
l = Symbol('l')
x = Symbol('x')
y = Symbol('y')

F = 5 * (x - l*(10*x+3))**2 + 6 * (y-l*(12*y-2))**2 + 3 * (x - l*(10*x+3)) - 2 * (y-l*(12*y-2)) + 3

F_derivative = F.diff(l)
F_derivative
```

##НСА

```python
def f_lamda(x, y):
  return (60*x - 48*y + 144*y**2 + 100*x**2 + 13) / (-576*y + 138 + 1728*y**2 + 1000*x**2 + 600*x)
```

```python
# Нахождение градиента функции
x = Symbol('x')
y = Symbol('y')

f_ = 5 * x**2 + 6 * y**2 + 3*x - 2*y + 3

dfdx = f_.diff(x)
dfdy = f_.diff(y)

print(dfdx, dfdy, sep='\n')

dfdx = lambdify(x, dfdx)
dfdy = lambdify(y, dfdy)
```

```python
n = 5
x0, y0 = 1, 1
df_NSA = pd.DataFrame(columns=['x', 'y', 'f', 'g1', 'g2', 'lamda'])

for i in range(n):
  fxy = f(x0, y0)
  lamda = f_lamda(x0, y0)
  g1 = dfdx(x0)
  g2 = dfdy(y0)

  # print(x0, y0, fxy, lamda, g1, g2)
  df_NSA.loc[len(df_NSA.index)] = [x0, y0, fxy, g1, g2, lamda]

  x0 = x0 - lamda * g1
  y0 = y0 - lamda * g2

df_NSA.loc[len(df_NSA.index)] = [x0, y0, fxy, np.nan, np.nan, np.nan]
```

```python
df_NSA
```

##ГДШ

```python
x0, y0 = 1, 1
df_GDS = pd.DataFrame(columns=['x', 'y', 'f', 'g1', 'g2', 'lamda'])

for _ in range(n):
  lamda = 0.618
  fxy = f(x0, y0)
  g1 = dfdx(x0)
  g2 = dfdy(y0)
  x1 = x0 - lamda * g1
  y1 = y0 - lamda * g2

  while fxy - f(x1, y1) < lamda / 2 * (g1**2 + g2**2):
    lamda /= 2
    x1 = x0 - lamda * g1
    y1 = y0 - lamda * g2

  # print(x0, y0, fxy, g1, g2, lamda)
  df_GDS.loc[len(df_GDS.index)] = [x0, y0, fxy, g1, g2, lamda]

  x0, y0 = x1, y1

```

```python
df_GDS
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_NSA.x, y=df_NSA.y, name='НСА'))
fig.add_trace(go.Scatter(x=df_GDS.x, y=df_GDS.y, name='ГДШ'))


fig.update_layout(xaxis_title="Значение x",
                  yaxis_title="Значение y",
                  xaxis=dict(scaleanchor="y", scaleratio=1))
```
